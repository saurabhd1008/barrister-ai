```python
# Install the Sentence Transformer library
!pip install --quiet --upgrade sentence-transformers

import os
import logging

# Suppress Hugging Face / Transformers logs
os.environ["TRANSFORMERS_VERBOSITY"] = "error"  # only show errors

logging.getLogger("transformers").setLevel(logging.ERROR)
logging.getLogger("sentence_transformers").setLevel(logging.ERROR)

from sentence_transformers import SentenceTransformer, util

# Load model silently
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
def get_embedding(text):
    return model.encode(text,convert_to_tensor=True)

def find_text_similarity(text1, text2):
    return round(float(util.pytorch_cos_sim(get_embedding(text1), get_embedding(text2))[0][0]), 4)

def find_embedding_similarity(emb1, emb2):
    return round(float(util.pytorch_cos_sim(emb1, emb2)[0][0]), 4)
```

# LLM Tools

```python
text = "I am a human being."
text_embedding = get_embedding(text)
print(len(text_embedding))
print(text_embedding)

# An embedding is a vector of numbers
# which captures the meaning of a given sentence in an abstract way.
# We cannot understand these numbers ourselves.
# But these are very helpful in many ML and AI applications.
# These embeddings basically form the input for all ML applications,
# and are a non-explainable substitute for feature engineering.
```

# Word Embedding Plots

```python
# Define word list
words = [
    "apple", "banana", "mango", "orange",
    "tiger", "lion", "cat", "dog",
    "car", "truck", "bicycle", "train",
    "happy", "sad", "angry", "excited"
]

# Define categories (same order as words)
categories = [
    "fruit", "fruit", "fruit", "fruit",
    "animal", "animal", "animal", "animal",
    "vehicle", "vehicle", "vehicle", "vehicle",
    "emotion", "emotion", "emotion", "emotion"
]
```

```python
# Get embeddings
embeddings = model.encode(words)
```

```python
# Reduce to 2D using PCA
# from sklearn.decomposition import PCA
# pca = PCA(n_components=2)
# points = pca.fit_transform(embeddings)

# # Use TSNE
from sklearn.manifold import TSNE
points = TSNE(n_components=2, perplexity=5, random_state=42).fit_transform(embeddings)
```

```python
# Create a DataFrame for Plotly
import pandas as pd
df = pd.DataFrame({
    'word': words,
    'category': categories,
    'x': points[:,0],
    'y': points[:,1]
})
```

```python
import plotly.express as px

# Interactive scatter plot
fig = px.scatter(
    df,
    x='x',
    y='y',
    color='category',
    text='word',
    title="🧩 Word Embeddings in 2D — Similar Words Stick Together!",
    color_discrete_sequence=px.colors.qualitative.Pastel
)

# Style and show
fig.update_traces(textposition='top center', marker=dict(size=12, opacity=0.8))
fig.update_layout(
    width=800,
    height=600,
    showlegend=True,
    template='plotly_white'
)

fig.show()
```
